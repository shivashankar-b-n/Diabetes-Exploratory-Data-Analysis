```python
# Count zero values in important medical columns

columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in columns:
    print(f"{col}: {(df[col] == 0).sum()}")
```

```python
import numpy as np

columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df[columns] = df[columns].replace(0, np.nan)
```

```python
df.isnull().sum()
```

```python
for col in columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.to_csv("diabetes_cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.countplot(x='Outcome', data=df, palette='Set2')

plt.title("Count Plot of Diabetes Outcome")
plt.xlabel("Outcome (0 = Non-Diabetic, 1 = Diabetic)")
plt.ylabel("Count")

plt.show()

# Explanation
print("Explanation:")
print("This graph shows the number of diabetic and non-diabetic patients.")
print("Most patients belong to the non-diabetic class.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

# Explanation
print("Explanation:")
print("This graph shows the distribution of patients' ages.")
print("Most patients are between 20 and 40 years old.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.boxplot(y=df['BMI'], color='lightgreen')

plt.title("Box Plot of BMI")
plt.ylabel("BMI")

plt.show()

# Explanation
print("Explanation:")
print("This graph helps identify outliers in BMI values.")
print("Some patients have unusually high BMI values.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))

sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()

# Explanation
print("Explanation:")
print("This heatmap shows the relationship between all numerical features.")
print("Darker colors indicate stronger positive or negative correlations.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.scatterplot(x='Glucose', y='BMI', data=df, color='blue')

plt.title("Scatter Plot of Glucose vs BMI")
plt.xlabel("Glucose")
plt.ylabel("BMI")

plt.show()

# Explanation
print("Explanation:")
print("This graph shows the relationship between Glucose and BMI.")
print("Higher glucose values are generally associated with higher BMI.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df, hue='Outcome')

plt.show()

# Explanation
print("Explanation:")
print("This graph shows the relationship between all numerical features.")
print("It helps identify patterns and differences between diabetic and non-diabetic patients.")
```
